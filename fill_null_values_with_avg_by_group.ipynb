```python
import pandas as pd
import numpy as np
```

```python
dummy = np.random.random((30,2))
```

```python
df = pd.DataFrame(dummy, columns= ['value', 'class'])
```

```python
def classify_data(val):
    if val > 0.3:
        return 1
    return 0
```

```python
def sweep_data(val):
    if val > 0.7:
        return np.NAN
    return val
```

```python
df['class'] = df['class'].apply(lambda x : classify_data(x))
```

```python
df['value'] = df['value'].apply(lambda x : sweep_data(x))
```

```python
df
```

```python
classes = df['class'].unique()
```

```python
def fill_null_values(data, mean):
    if data > 0:
        return data
    return mean
```

```python
filled_df_list = []
for x in classes:
    mean = df[ df['class'] == x ]['value'].mean()
    print(mean)
    temp_df = df[ df['class'] == x ]
    temp_df['value'] = temp_df['value'].apply(lambda x : fill_null_values(x, mean))
    print(temp_df)
    filled_df_list.append(temp_df)
```

```python
final_df = pd.concat(filled_df_list)
```
